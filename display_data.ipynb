```python
import altair as alt

from police_data import get_crime_data

neighbourhood, crimes = get_crime_data()
```

```python
month_crimes = crimes["month"].value_counts().reset_index()

alt.Chart(month_crimes).mark_line(point=True, strokeWidth=3).encode(
    x=alt.X('month', title='Month'),
    y=alt.Y('count', title='Number of Crimes'),
    color=alt.value('teal'),
    tooltip=[
        alt.Tooltip('month', title='Month'),
        alt.Tooltip('count',
                    title='Number of Crimes')
    ]
).properties(
    width=700,
    height=300,
    title=f'Number of Crimes per Month in 2024 in {neighbourhood.title()}'
).configure_point(
    size=100
)
```

```python
locations = crimes["location_name"].value_counts().reset_index()

alt.Chart(locations.head(15)).mark_bar().encode(
    x=alt.X('location_name', title='Location Type/Name'),
    y=alt.Y('count', title='Number of Crimes'),
    color=alt.Color('count', title='Number of Crimes'),
    tooltip=[
        alt.Tooltip('location_name', title='Location Type/Name'),
        alt.Tooltip(
            'count', title='Number of Crimes')
    ]
).properties(
    width=600,
    height=500,
    title=f'Top 15 Most Dangerous Areas in {neighbourhood.title()}'
)
```

```python
categories = crimes["category"].value_counts().reset_index()

alt.Chart(categories).mark_bar().encode(
    x=alt.X('category', title='Crime Category'),
    y=alt.Y('count', title='Number of Crimes'),
    color=alt.Color('count', title='Number of Crimes'),
    tooltip=[
        alt.Tooltip('category', title='Crime Category'),
        alt.Tooltip(
            'count', title='Number of Crimes')
    ]
).properties(
    width=600,
    height=500,
    title=f'Most Common Crimes in {neighbourhood.title()}'
)
```
